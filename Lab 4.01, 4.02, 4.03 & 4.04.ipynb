```python
def snake_case_columns(dataframe):
    
    snake_columns = list()
    
    for column in dataframe.columns:
        column = column.replace(' ','_')
        column = column.lower()
        snake_columns.append(column)
    
    dataframe.columns = snake_columns
    
    return dataframe

customer_df = snake_case_columns(customer_df)
```

```python
#there are many unique values, so it would be better to separate the date into days, months and years
```

```python
def year_month_day(customer_df, effective_to_date):

    year = customer_df[effective_to_date].dt.year
    month = customer_df[effective_to_date].dt.month
    day = customer_df[effective_to_date].dt.day

    customer_df['year'] = year
    customer_df['month'] = month
    customer_df['day'] = day

    return customer_df

customer_df = year_month_day(customer_df, 'effective_to_date')
```

```python
numerical_df = numerical_df[(numerical_df['customer_lifetime_value']>lower_limit) & (numerical_df['customer_lifetime_value']<upper_limit)]
sns.kdeplot(numerical_df['customer_lifetime_value'])
plt.show()
```

```python
#Monthly premium


iqr = np.percentile(numerical_df['monthly_premium_auto'],75) - np.percentile(numerical_df['monthly_premium_auto'],25)
upper_limit = np.percentile(numerical_df['monthly_premium_auto'],75) + 1.5*iqr
lower_limit = np.percentile(numerical_df['monthly_premium_auto'],25) - 1.5*iqr
```
